```python
time_series = timeseries_df.groupby('Date')['Amount'].sum()

plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
plt.plot(time_series.index, time_series.values, marker='o', linestyle='-')
plt.title('Total Amount Received Over Time')
plt.xlabel('Date')
plt.ylabel('Total Amount Received')
plt.grid(True)

plt.xticks(rotation=45)

plt.show()
```
