## Ratio by activity sector

```python
import re
# opp_df['Activity sector'] = opp_df['Activity sector'].replace(['SECTEUR PUBLIC'], 'PUBLIC SECTOR')
opp_df['Activity sector'].fillna('OTHER', inplace=True)
# opp_df['Activity sector'] = opp_df['Activity sector'].replace([pd.na()], 'OTHER')
opp_df["activity_sector_agg"] = opp_df["Activity sector"].apply(lambda x:' '.join(re.findall(r'\b[A-Z]+\b', x)) if not pd.isna(x) else x)
# opp_df["activity_sector_agg"] = opp_df["Activity sector"].apply(lambda x:x.split("-")[0] if not pd.isna(x) else x)
opp_df['activity_sector_agg'] = opp_df['activity_sector_agg'].replace(['COMPUTING TELECOM E'], 'COMPUTING TELECOM')
opp_df['activity_sector_agg'] = opp_df['activity_sector_agg'].replace(['SECTEUR PUBLIC'], 'PUBLIC SECTOR')
opp_df
```

```python
opp_df["real_money"] = opp_df.apply(lambda row:row["Estimated amount"] if row.Status == "Won" else 0, axis=1)
# full_df["real_money"] = full_df.apply(lambda row:row["Estimated amount"] if row.Status == "Won" else 0, axis=1)
# opp_df[""]
opp_df["total_estimated_money"] = opp_df.apply(lambda row:row["Estimated amount"] if row.Status in ["Won", "Lost"] else 0, axis=1)
opp_df["ratio"] = opp_df["real_money"]
ratio_df = opp_df.groupby(["activity_sector_agg"])["Business Account", "Opportunity ID", "real_money", "total_estimated_money"].agg({"Business Account" : lambda x: x.nunique(), "Opportunity ID":lambda x: x.nunique(), "real_money":"sum", "total_estimated_money":"sum"}) # , "City_x"
ratio_df = ratio_df.rename(columns={'Business Account': 'n_disctinct_business_accounts', 'Opportunity ID': 'n_distinct_opportunities'})
ratio_df["success_ratio"] = ratio_df["real_money"]/ratio_df["total_estimated_money"]
ratio_df["avg_donation_per_opportunity"] = ratio_df["real_money"]/ratio_df["n_distinct_opportunities"]
ratio_df.sort_values(by="success_ratio", ascending=False)
```
